```python
import pandas as pd 
from bs4 import BeautifulSoup
from sqlalchemy import create_engine
```

### Importing and Cleaning USA Used Cars Data

```python
use_car_test = pd.read_csv("data/USA_cars_datasets.csv")
use_car_test.head()

```

```python
clean_used_cars = use_car_test.loc[use_car_test['title_status'] == "clean vehicle"]
clean_used_cars = clean_used_cars.drop(columns = ['Unnamed: 0','color','vin','lot','state','country','condition'])
clean_used_cars = clean_used_cars.rename(columns={"price":"current_price"})



clean_used_cars
```

```python
clean_used_cars.reset_index(inplace = True)
clean_used_cars.rename(columns={"brand":"make"},inplace = True)
```

```python
clean_used_cars
```

### Importing and Cleaning New Car Data 

```python
new_car_test = pd.read_csv("data/data.csv")
new_car_test.head()
```

```python
new_cars = new_car_test.rename(columns = {"Make":"make","Model":"model","Year":"year","Engine Fuel Type":"engine_fuel_type","Engine HP":"engine_hp","Engine Cylinders":"engine_cylinders","Transmission Type":"transmission_type","Driven_Wheels":"driven_wheels","Number of Doors":"number_of_doors","Market Category":"market_category","Vehicle Size":"vehicle_size","Vehicle Style":"vehicle_style","highway MPG":"highway_mpg","city mpg":"city_mpg","Popularity":"popularity","MSRP":"msrp"})

new_cars.info()
```

```python
new_cars.reset_index(inplace =True)

new_cars

new_cars.rename(columns = {"index":"index_col"},inplace = True)

```

```python
new_cars.drop(columns= ['level_0'],inplace = True)
```

```python
new_cars
```

### Connect to local database

```python
rds_connection_string = "postgres:postgres@localhost:5432/car_data"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for Tables

```python
engine.table_names()
```

### Use Pandas to load dataframe into database

```python
new_cars.to_sql(name='cars_db',con = engine, if_exists='append',index=False)
```

```python
clean_used_cars.to_sql(name='used_cars_db',con = engine, if_exists = 'append',index=False)
```

```python
pd.read_sql_query('select * from cars_db',con=engine).head()
```

```python
pd.read_sql_query('select * from used_cars_db',con = engine).head()
```
